#### Writing and Reading Twitter JSON

```python
# Twitter may have deleted the twitter_id-- creating 2 different lists 
yes_tweets = []
no_tweets = []


#Tweet_id from twitter_archive column
t_a = twitter_archive['tweet_id']

#Saving in two different lists 
for tweet_id in t_a:
    try: 
        yes_tweets.append(api.get_status(tweet_id))   #Seperate tweets with no tweet_id and tweets without tweet_id 
    except Exception as e: 
        no_tweets.append(tweet_id)
```

source: https://stackoverflow.com/questions/3768895/how-to-make-a-class-json-serializable

source: https://stackoverflow.com/questions/50428033/converting-list-into-dictionary-with-index-as-key

```python
#import json and get append json_tweet in dicts
import json
dicts = []
for json_tweet in yes_tweets :
    dicts.append(json_tweet._json)
```

```python
#write dict in txt file
with open('tweet_json.txt', 'w') as file:
        file.write(json.dumps(dicts, indent=4))
```

```python
#Now Create a DataFrame from the tweet_json.txt file:
another_list = []
with open('tweet_json.txt', encoding='utf-8') as json_file:  
    data = json.load(json_file)
    for each_dict in data:
        tweet_id = each_dict['id']
        favorite_count = each_dict['favorite_count']
        retweet_count = each_dict['retweet_count']
        another_list.append({'tweet_id': str(tweet_id),
                             'favorite_count': int(favorite_count),
                             'retweet_count': int(retweet_count),
                            })
        #print(my_demo_list)
        tweet_json = pd.DataFrame(another_list, columns = 
                                  ['tweet_id', 
                                   'favorite_count', 'retweet_count', 
                                   ])
```

#### Replace:

Twitter Archive table:
- 1 . Remove non-recoverable Gofundme from extended URLS

#####  1 &  2.Code 

```python
# Remove non-recoverable Gofundme from extended 
ta_clean = ta_clean[ta_clean.expanded_urls != 'https://gofundme.com/ydvmve-surgery-for-jax,ht...']
```

#### 3 .Define

- 3 . Replace articles (such as the, a, an) and few missing data (None) to '  ' from name column

```python
ta_clean['name'].tail(25)
```

source : https://stackoverflow.com/questions/23743460/replace-none-with-nan-in-pandas-dataframe

##### 3.Code

```python
#Replace "a, an, None, quite,such, the" to NAN

ta_clean['name'].replace('None', np.nan, inplace=True)
ta_clean['name'].replace('a', np.nan, inplace=True)
ta_clean['name'].replace('an', np.nan, inplace=True)
ta_clean['name'].replace('quite', np.nan, inplace=True)
ta_clean['name'].replace('the', np.nan, inplace=True)
ta_clean['name'].replace('such', np.nan, inplace=True)
```

##### 3 .Test

```python
#Verify if NAN are replacing them
ta_clean['name'].tail()
```

##### 6. Define
- 6 . Make one column for doggo, floofer, pupper, and puppo

#### 6.Code

source: https://stackoverflow.com/questions/39291499/how-to-concatenate-multiple-column-values-into-a-single-column-in-panda-datafram

```python
#Replace all the None to blank
ta_clean.doggo = ta_clean.doggo.replace('None',' ')
ta_clean.floofer = ta_clean.floofer.replace('None',' ')
ta_clean.pupper = ta_clean.pupper.replace('None',' ')
ta_clean.puppo = ta_clean.puppo.replace('None',' ')
```

```python
#Merge all four columns to one column 'dog'. 
ta_clean['dog']= ta_clean['doggo'].astype('object')+' '+ta_clean['floofer'].astype('object')+' '+ta_clean['pupper']+' '+ta_clean['puppo'].astype('object')
```

```python
#Drop doggo, floofer, pupper, and puppo
ta_clean.drop(['doggo'], axis =1, inplace = True)
ta_clean.drop(['floofer'], axis =1, inplace = True)
ta_clean.drop(['pupper'], axis =1, inplace = True)
ta_clean.drop(['puppo'], axis =1, inplace = True)
```
